```python
#
# 1-D and 2-D QFT Examples
#
# Alexis A. Gomez, Dr. Artyom Grigoryan,
# UTSA ECE, March 11, 2023
#
# Reference: 
#   http://dx.doi.org/10.1117/12.3013940
#   Quantum Image Processing: A Mathematical Toolbox by Artyom M. Grigoryan and Sos Agaian

# Import to Draw Circuit and Display Version
from qiskit import __version__
from qiskit.visualization import array_to_latex
from qiskit.quantum_info.operators import Operator

# Import for QFT algorithm
import numpy as np
from qu_algorithms import qft

# Custom Draw Settings
draw_settings = {'output': "mpl", 'style': "bw", 'scale': 1.5, 'fold': 12}

# draw_circuit` function is designed for visualizing and optionally saving a representation of a quantum circuit. It
# takes a quantum circuit object (`circuit`), a string for naming the file (`name`, defaulting to 'circuit1'),
# and two boolean flags (`getArray` and `savefig`) as inputs. The function first creates a visual drawing of the
# circuit using predefined drawing settings (`draw_settings`). If `savefig` is set to True, this visual
# representation is saved as an SVG file with the provided name. If `getArray` is True, the function also converts
# the circuit (with reversed bits) into a LaTeX array format using the `array_to_latex` function applied to the
# circuit's operator. Depending on the values of `getArray` and `savefig`, the function returns either the drawn
# circuit, both the drawn circuit and its LaTeX array representation, or just the drawn circuit.
def draw_circuit(circuit, name='circuit1', getArray=False, savefig=False):
    circuitDrawn = circuit.draw(**draw_settings)
    if savefig:
        circuitDrawn.savefig(str(name) + '.svg')
    if getArray:
        array = array_to_latex(Operator(circuit.reverse_bits()))
        return circuitDrawn, array

    return circuitDrawn
```

# 1-D N-qubit Examples
This sections covers how to use the qu_algorithms.qft.qft module for the N-qubit QFT.


## 4-qubit QFT with Random Input
State Preparation is done using the Digital signal-induced Heap Transform (DsiHT).

```python
# Create Input as a 1-D Vector (1x16)
ImageInput = np.random.randint(100,255,size=(1,16))

# Retrieve 4-qubit QFT circuit
qc = qft.qft(ImageInput)

# Display 4-qubit QFT. Use 'reverse_bits' to compensate for little-endian notation.
display(draw_circuit(qc.reverse_bits()))

# Verify Results with Classical Results
ver = qft.verify_qft(ImageInput, qc, print_bool=True)
```

## 4-qubit QFT without any Input

```python
# Retrieve 4-qubit QFT circuit without any input
qc = qft.qft(4)

# Display 4-qubit QFT. Use 'reverse_bits' to compensate for little-endian notation.
display(draw_circuit(qc.reverse_bits()))
```

# 2-D NxN-qubit Examples
This sections covers how to use the qu_algorithms.qft.qft2 module for the NxN-qubit QFT.


## 4x4-qubit QFT with Random Input
State Preparation is done using the Digital signal-induced Heap Transform (DsiHT).

```python
# Create Input as a 1-D Vector (1x256 for 16x16 Image)
ImageInput = np.random.randint(100,255,size=(1,16*16))

# Retrieve 4x4-qubit QFT circuit
qc = qft.qft2(ImageInput)

# Display 4x4-qubit QFT. Use 'reverse_bits' to compensate for little-endian notation.
display(draw_circuit(qc.reverse_bits()))

# Verify Results with Classical Results
ver2 = qft.verify_qft2(ImageInput, qc, print_bool=True)
```

## 4x4-qubit QFT without any Input

```python
# Retrieve 4x4-qubit QFT circuit without any input
qc = qft.qft2(4)

# Display 4x4-qubit QFT. Use 'reverse_bits' to compensate for little-endian notation.
display(draw_circuit(qc.reverse_bits()))
```
